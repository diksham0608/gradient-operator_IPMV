```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/diksha.jfif', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded
if image is None:
    print("Error loading image.")
    exit()

# Apply Sobel operator to find the gradient in both X and Y direction
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in X direction
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in Y direction

# Compute the magnitude of the gradient
gradient_magnitude = cv2.magnitude(sobel_x, sobel_y)

# Normalize the result to 8-bit values
gradient_magnitude = np.uint8(cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX))

# Display the original image and gradient magnitude
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Gradient Magnitude')
plt.imshow(gradient_magnitude, cmap='gray')

plt.show()
```
